```python
import pandas as pd 
import numpy as np
```

```python
credits = pd.read_csv(r"C:\Users\bkish\csv\tmdb_5000_credits.csv") 
movies_df = pd.read_csv(r"C:\Users\bkish\csv\tmdb_5000_movies.csv") 
```

```python
credits.head() 
movies_df.head()
```

```python
print("Credits:",credits.shape) 
print("Movies Dataframe:",movies_df.shape)
```

```python
credits_column_renamed = credits.rename(index=str, 
columns={"movie_id": "id"}) 
movies_df_merge = movies_df.merge(credits_column_renamed, on='id') 
movies_df_merge.head()
```

```python
movies_cleaned_df = movies_df_merge.drop(columns=['homepage', 'title_x', 
'title_y', 'status','production_countries']) 
movies_cleaned_df.head() 
movies_cleaned_df.info() 
# Calculate all the components based on the above 
v=movies_cleaned_df['vote_count'] 
R=movies_cleaned_df['vote_average'] 
C=movies_cleaned_df['vote_average'].mean() 
m=movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_average']=((R*v)+ 
(C*m))/(v+m) 
movies_cleaned_df.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python

movies_df['weighted_average'] = (movies_df['vote_average'] * movies_df['vote_count']) / (movies_df['vote_count'] + 1)

movie_sorted_ranking = movies_df.sort_values('weighted_average', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(
    x=movie_sorted_ranking['weighted_average'].head(10),
    y=movie_sorted_ranking['original_title'].head(10)
)
plt.xlim(4, 10)
plt.title('Best Movies by Average Votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling=MinMaxScaler() 
movie_scaled_df=scaling.fit_transform(movies_cleaned_df
 [['weighted_average','popularity']])
```

```python
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity']) 
movie_normalized_df.head() 
```

```python
movies_cleaned_df[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_average'] * 0.5 + movies_cleaned_df['normalized_popularity'] * 0.5

movies_scored_df = movies_cleaned_df.sort_values(['score'], ascending=False)

movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
scored_df = movies_cleaned_df.sort_values('score', 
ascending=False) 
```
